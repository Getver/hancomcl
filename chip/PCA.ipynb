```r
library(ggplot2)
library(reshape2)
library(dplyr)
library(ggrepel)
# library(tidyverse)
```

```r
pca <- read.table('/disk0/sm/chip/01_KBA/1031/06_pca/all_pca.eigenvec', sep='\t', header=F)
```

```r
dim(pca)
head(pca, 2)
```

```r
p <- ggplot(pca, aes(x=V3, y=V4)) + 
  # coord_cartesian(xlim = c(0.5, 1.0), ylim = c(50, 100)) +
  geom_point(size=0.5, color='#0180c9') + 
    geom_text_repel(
    data = subset(pca, abs(V3) > 0.05 | abs(V4) > 0.05),
    aes(label = V2),
    size = 3,
    color = 'black',
    max.overlaps = 50
  ) +
  # scale_x_continuous(breaks = seq(0.5, 1.0, 0.1)) +
  # scale_y_continuous(breaks = seq(50, 100, 5)) +
  labs(title='PCA',
      x = 'pc1', 
      y = 'pc2') +
  theme_bw() +
  theme(
    axis.text.x = element_text(size=15),
    axis.text.y = element_text(size=15),
    plot.title = element_text(hjust = 0.5,size=20,face='bold'),
    axis.title.x = element_text(size=18),
    axis.title.y = element_text(size=18),
    panel.grid.major.x = element_line(linetype = "dotted", colour = "gray80"), 
    panel.grid.major.y = element_line(linetype = "dotted", colour = "gray80"),
    panel.grid.minor.x = element_blank(),
    panel.grid.minor.y = element_blank(),
    legend.position = 'right',
    )
  # geom_vline(xintercept=0.82, color='#E74646') + 
  # geom_hline(yintercept=96, color='#E74646') +
  # scale_color_manual(values=adjustcolor(col, alpha.f = 0.3))
ggsave(p, file='/disk0/sm/chip/01_KBA/1031/06_pca/PCA.png', width=5000, height=5000, units='px', dpi=500)
```
