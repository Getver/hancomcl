```r
library(ggplot2)
library(reshape2)
library(dplyr)
library(ggrepel)
# library(tidyverse)
```

```r
sm_color = c('#450693', '#FF3F7F', '#0046FF', '#57564F', '#08D9D6', '#8C00FF', '#FF2E63', '#FFC400', '#280A3E')
```

```r
key <- 'all1'
folder <- paste0('/disk0/sm/chip/01_KBA/', key)
data <- read.table(paste0(folder, '/QC_TABLE.txt'), sep='\t', header=TRUE)
```

```r
data$plate <- sub(".*_B_(\\d+)_.*", "\\1", data$cel_files)
```

```r
het <- read.table(paste0(folder, '/03_vcf/all1_plink_het.het')) %>% select(V2, V6)
H <- merge(data, het, by.x='cel_files', by.y='V2', all=TRUE)
head(H, 2)
```

```r
## DQC

p <- ggplot(data, aes(x=reorder(cel_files, DQC), y=DQC*100)) + 
  coord_cartesian(ylim = c(80, 100)) +
  geom_point(size=0.2, color=sm_color[1]) + 
  # scale_x_continuous(breaks = seq(0.5, 1.0, 0.1)) +
  scale_y_continuous(breaks = seq(80, 100, 2)) +
  labs(title='DQC', x='sample(명)', y='dqc(%)') +
  theme_bw() +
  theme(
    axis.text.x = element_blank(),
    axis.text.y = element_text(size=10),
    plot.title = element_text(hjust = 0.5,size=12,face='bold'),
    axis.title.x = element_text(size=10),
    axis.title.y = element_text(size=10),
    axis.ticks.x = element_blank(),
    panel.grid.major.x = element_blank(),
    panel.grid.major.y = element_line(linetype = "dotted", colour = "gray80"),
    panel.grid.minor.x = element_blank(),
    panel.grid.minor.y = element_blank(),
    legend.position = 'right',
    ) +
  # geom_vline(xintercept=82, color='#E74646') + 
  geom_hline(yintercept=82, color='#E74646')
  # scale_color_manual(values=sm_color[1])

ggsave(p, file='/disk0/sm/chip/plot/DQC_point.png', width=4000, height=2000, units='px', dpi=500)


p <- ggplot(data, aes(x=plate, y=DQC*100)) + 
  coord_cartesian(ylim = c(80, 100)) +
  geom_boxplot(size=0.2, fill=sm_color[1], outliers=FALSE) + 
  # scale_x_continuous(breaks = seq(0.5, 1.0, 0.1)) +
  scale_y_continuous(breaks = seq(80, 100, 2)) +
  labs(title='DQC', x='plate', y='dqc(%)') +
  theme_bw() +
  theme(
    axis.text.x = element_blank(),
    axis.text.y = element_text(size=10),
    plot.title = element_text(hjust = 0.5,size=12,face='bold'),
    axis.title.x = element_text(size=10),
    axis.title.y = element_text(size=10),
    axis.ticks.x = element_blank(),
    panel.grid.major.x = element_blank(),
    panel.grid.major.y = element_line(linetype = "dotted", colour = "gray80"),
    panel.grid.minor.x = element_blank(),
    panel.grid.minor.y = element_blank(),
    legend.position = 'right',
    ) +
  # geom_vline(xintercept=82, color='#E74646') + 
  geom_hline(yintercept=82, color='#E74646')
  # scale_color_manual(values=sm_color[1])

ggsave(p, file='/disk0/sm/chip/plot/DQC_box.png', width=4000, height=2000, units='px', dpi=500)
```
